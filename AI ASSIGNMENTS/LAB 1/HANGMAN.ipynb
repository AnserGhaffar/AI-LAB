```python
import random

class Hangman:
    def __init__(self):
        self.words = ["python", "java", "kotlin", "javascript", "hangman", "challenge", "random", "computer", "programming"]
        self.word = random.choice(self.words)
        self.guessed_word = ["_"] * len(self.word)
        self.guessed_letters = set()
        self.attempts = 6

    def display_hangman(self):
        stages = [
            """
               --------
               |      |
               |      O
               |     \\|/
               |      |
               |     / \\
               -
            """,
            """
               --------
               |      |
               |      O
               |     \\|/
               |      |
               |     / 
               -
            """,
            """
               --------
               |      |
               |      O
               |     \\|/
               |      |
               |      
               -
            """,
            """
               --------
               |      |
               |      O
               |     \\|
               |      |
               |     
               -
            """,
            """
               --------
               |      |
               |      O
               |      |
               |      |
               |     
               -
            """,
            """
               --------
               |      |
               |      O
               |    
               |      
               |     
               -
            """,
            """
               --------
               |      |
               |      
               |    
               |      
               |     
               -
            """
        ]
        return stages[self.attempts]

    def make_guess(self, guess):
        if guess in self.guessed_letters:
            print("You already guessed that letter.")
            return False

        self.guessed_letters.add(guess)

        if guess in self.word:
            for i, letter in enumerate(self.word):
                if letter == guess:
                    self.guessed_word[i] = letter
            return True
        else:
            self.attempts -= 1
            return False

    def play(self):
        print("Welcome to Hangman!")

        while self.attempts > 0:
            print(self.display_hangman())
            print("Word to guess: ", " ".join(self.guessed_word))
            guess = input("Guess a letter: ").lower()

            if self.make_guess(guess):
                if "_" not in self.guessed_word:
                    print("\nCongratulations! You've guessed the word:", self.word)
                    break
            else:
                print(f"Wrong guess. You have {self.attempts} attempts left.")

            if self.attempts == 0:
                print(self.display_hangman())
                print("\nSorry, you've run out of attempts. The word was:", self.word)

if __name__ == "__main__":
    game = Hangman()
    game.play()
```
